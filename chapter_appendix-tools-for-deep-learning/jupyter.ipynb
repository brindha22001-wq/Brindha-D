### Decision Tree Classifier

A Decision Tree Classifier is a non-parametric supervised learning method used for classification and regression. The goal is to create a model that predicts the value of a target variable by learning simple decision rules inferred from the data features. Here's how to create one:

1.  **Import the model:** Import `DecisionTreeClassifier` from `sklearn.tree`.
2.  **Instantiate the model:** Create an instance of the `DecisionTreeClassifier`.
3.  **Train the model:** Use the `fit` method to train the model on your training data (`x_train`, `y_train`).
4.  **Make predictions:** Use the `predict` method to make predictions on the test data (`x_test`).
5.  **Evaluate the model:** Calculate accuracy and other classification metrics using `accuracy_score`, `confusion_matrix`, and `classification_report` from `sklearn.metrics`.

```python
# Import the Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

# Create an instance of the Decision Tree Classifier
dtree_model = DecisionTreeClassifier(random_state=42) # Added random_state for reproducibility

# Train the model on the scaled training data
dtree_model.fit(x_train, y_train)

print("Decision Tree Classifier trained successfully!")
```

```python
# Make predictions on the scaled test data
y_pred_dtree = dtree_model.predict(x_test)

print("Predictions made successfully!")
```

### K-Nearest Neighbors (KNN) Classifier

K-Nearest Neighbors is a non-parametric, lazy learning algorithm used for both classification and regression. In classification, the output is a class membership. An object is classified by a majority vote of its neighbors, with the object being assigned to the class most common among its k-nearest neighbors.

Here's how to use the KNN Classifier:

1.  **Instantiate the model:** We've already done this in a previous step, where `knn` was created as an instance of `KNeighborsClassifier`.
2.  **Train the model:** Use the `fit` method to train the model on your scaled training data (`x_train`, `y_train`).
3.  **Make predictions:** Use the `predict` method to make predictions on the scaled test data (`x_test`).
4.  **Evaluate the model:** Calculate accuracy, confusion matrix, and classification report using `accuracy_score`, `confusion_matrix`, and `classification_report` from `sklearn.metrics`.

```python
# Train the KNN model on the scaled training data
knn.fit(x_train, y_train)

print("K-Nearest Neighbors Classifier trained successfully!")
```

```python
# Make predictions on the scaled test data
y_pred_knn = knn.predict(x_test)

print("Predictions made successfully with KNN!")
```

```python
# Evaluate the K-Nearest Neighbors Classifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("K-Nearest Neighbors Classifier Performance:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_knn):.6f}")
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred_knn))
```
